# Embedding — Mengubah Teks Menjadi Vector

```python
from langchain_google_genai import GoogleGenerativeAIEmbeddings

embedding_model = GoogleGenerativeAIEmbeddings(google_api_key=api_key, model="gemini-embedding-2", output_dimensionality=768)
```

# Vector Database — ChromaDB

```python
from langchain_community.vectorstores import Chroma

# Sematkan setiap chunk dan muat ke dalam chromadb
db = Chroma.from_documents(chunks, embedding_model, persist_directory="./chroma_db_")

# Menyimpan perubahan ke disk
db.persist()
```

```python
# membuka koneksi ke ChromaDB yang sudah tersimpan
# mengatur koneksi untuk menghubungkan ke ChromaDB
db_connection = Chroma(persist_directory="./chroma_db_", embedding_function=embedding_model)
```

```python
# Mengonversi koneksi Chroma menjadi objek Retriever untuk pencarian dokumen berbasis vektor
retriever = db_connection.as_retriever(search_kwargs={"k": 10})

print(type(retriever))
```

### Mengambil Hasil Relevan Beserta Score
Kita menggunakan `similarity_search_with_score` untuk mendapatkan nilai jarak (distance). Semakin kecil nilainya, semakin relevan dokumen tersebut.

```python
query = "Jelaskan tentang komodo?"
# Mengambil top 1 hasil paling relevan
results_with_score = db_connection.similarity_search_with_score(query, k=1)

if results_with_score:
    doc, score = results_with_score[0]
    print(f"Relevance Score (Distance): {score}")
    print("\nContent:")
    display(md(doc.page_content))
else:
    print("Tidak ada dokumen yang ditemukan.")
```

### Mengambil Beberapa Hasil Relevan Berdasarkan Score
Kita dapat mengatur parameter `k` untuk mengambil lebih dari satu dokumen dan melihat skor relevansi masing-masing.

```python
query = "Jelaskan tentang komodo?"
# Mengambil top 3 hasil paling relevan
results_with_score = db_connection.similarity_search_with_score(query, k=3)

print(f"Ditemukan {len(results_with_score)} dokumen relevan:\n")

for i, (doc, score) in enumerate(results_with_score):
    print(f"--- Hasil Ke-{i+1} ---")
    print(f"Relevance Score (Distance): {score}")
    display(md(doc.page_content))
    print("\n")
```

### Filter Berdasarkan Threshold Skor
Kita hanya akan menampilkan hasil yang memiliki skor jarak (distance) di bawah ambang batas tertentu agar lebih akurat.

```python
query = "Jelaskan tentang komodo?"
threshold = 0.5  # Sesuaikan nilai ini (semakin kecil semakin ketat)

results_with_score = db_connection.similarity_search_with_score(query, k=5)

# Filter hasil yang sesuai threshold
filtered_results = [(doc, score) for doc, score in results_with_score if score < threshold]

print(f"Ditemukan {len(filtered_results)} dokumen yang memenuhi threshold (< {threshold}):\n")

for i, (doc, score) in enumerate(filtered_results):
    print(f"--- Hasil Ke-{i+1} ---")
    print(f"Relevance Score: {score}")
    display(md(doc.page_content))
    print("\n")
```

# Prompt Template

```python
from langchain_core.messages import SystemMessage
from langchain_core.prompts import ChatPromptTemplate, HumanMessagePromptTemplate
```

```python
# Membuat template pesan untuk sistem dan pesan pengguna
chat_template = ChatPromptTemplate.from_messages([
    # System Message Prompt Template
    SystemMessage(content="""Anda adalah AI yang dapat menjawab pertanyaan berdasarkan konteks dan pertanyaan dari user.
                 Anda harus menjawab pertanyaan user, berdasarkan konteks"""),

    # Human Message Prompt Template
    HumanMessagePromptTemplate.from_template("""Jawab pertanyaan berikut berdasarkan konteks.
    konteks: {context}
    pertanyaan: {question}
    jawaban: """)
])
```

###Output Parser & Format Docs

```python
from langchain_core.output_parsers import StrOutputParser

output_parser = StrOutputParser()
```

```python
def format_docs(docs):
    return "\n\n".join(doc.page_content for doc in docs)
```

## RAG Chain — Menggabungkan Semua Komponen

```python
from langchain_core.runnables import RunnablePassthrough


rag_chain = (
    {"context": retriever | format_docs, "question": RunnablePassthrough()}
    | chat_template
    | chat_model
    | output_parser
)
```

## Chatbot dengan Gemini

```python
import subprocess
import time

def run_streamlit(filename, port=8501):
    # Kill SEMUA proses streamlit, bukan hanya yang kita spawn
    subprocess.run(["pkill", "-f", "streamlit"], capture_output=True)

    # Force-free port kalau masih ada yang nempel
    subprocess.run(["fuser", "-k", f"{port}/tcp"], capture_output=True)

    # Tutup semua tunnel ngrok
    ngrok.kill()

    # Tunggu port benar-benar bebas
    time.sleep(3)

    proc = subprocess.Popen(
        [
            "streamlit", "run", filename,
            "--server.headless=true",
            "--server.port", str(port),
            "--server.enableCORS=false",
        ],
        stdout=subprocess.DEVNULL,
        stderr=subprocess.DEVNULL
    )

    time.sleep(3)

    public_url = ngrok.connect(port)
    print(f"Streamlit berjalan: {public_url}")

    return proc
```

```python
%%writefile streamlit_chat_app.py
import streamlit as st
import google.generativeai as genai
from langchain_google_genai import ChatGoogleGenerativeAI, GoogleGenerativeAIEmbeddings
from langchain_community.vectorstores import Chroma
from langchain_core.output_parsers import StrOutputParser

st.set_page_config(page_title="Manual RAG Chat", layout="wide")
st.title("Chatbot Hewan Langka di Indonesia")
st.caption("Chatbot edukasi tsederhana menggunakan RAG - Google Gemini Flash")

# --- Initialization (Moved inside the script) ---
@st.cache_resource
def init_rag():
    # Use a hardcoded key or a text input for the standalone app
    # For this demo, we assume the user provides it via UI or it's passed via environment
    embedding_model = GoogleGenerativeAIEmbeddings(model="gemini-embedding-2", output_dimensionality=768)
    db_connection = Chroma(persist_directory="./chroma_db_", embedding_function=embedding_model)
    retriever = db_connection.as_retriever(search_kwargs={"k": 5})
    chat_model = ChatGoogleGenerativeAI(model="gemini-2.5-flash-lite")
    output_parser = StrOutputParser()
    return retriever, chat_model, output_parser

def format_docs(docs):
    return "\n\n".join(doc.page_content for doc in docs)

if "messages" not in st.session_state:
    st.session_state.messages = []

with st.sidebar:
    google_api_key = st.text_input("Google AI API Key", type="password")
    if st.button("Clear Chat"):
        st.session_state.messages = []
        st.rerun()

if not google_api_key:
    st.info("Masukkan API Key di sidebar.")
    st.stop()

# Configure API for the session
import os
os.environ["GOOGLE_API_KEY"] = google_api_key

# Load components
retriever, chat_model, output_parser = init_rag()

# Display History
for msg in st.session_state.messages:
    with st.chat_message(msg["role"]):
        st.markdown(msg["content"])

# Input logic
if prompt := st.chat_input("Tanya sesuatu..."):
    st.session_state.messages.append({"role": "user", "content": prompt})
    with st.chat_message("user"):
        st.markdown(prompt)

    with st.chat_message("assistant"):
        # 1. Retrieve
        with st.status("Mengambil data dari dokumen...") as status:
            docs = retriever.invoke(prompt)
            context_text = format_docs(docs)
            status.update(label="Konteks ditemukan!", state="complete")

        # 2. Manual Prompt Construction
        with st.spinner("Gemini sedang merespon..."):
            full_prompt = f"Jawablah berdasarkan konteks berikut:\n\n{context_text}\n\nPertanyaan: {prompt}"
            response = chat_model.invoke(full_prompt)
            answer = output_parser.invoke(response)

            st.markdown(answer)
            st.session_state.messages.append({"role": "assistant", "content": answer})
```
